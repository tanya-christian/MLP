```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([0,20])
y=np.array([25,200])
#showing line graph
plt.plot(x,y)
plt.show()
```

```python
#pointing marker
plt.plot(x,y,'o')
plt.show()
```

```python
x1=np.array([2,13,25,67])
y1=np.array([4,5,67,45])
plt.plot(x1,y1)
plt.show()
```

```python
plt.plot(x1,y1,marker='*')
plt.show()
```

```python
#marker in graph with color
plt.plot(x1,y1,marker='d',color='r')
plt.show()
```

```python
#marker adge color, marker size,marker face color
plt.plot(x1,y1,marker='d',ms=10,mec='r',mfc='y',color='b')
plt.show()
```

```python
#line style size and color
plt.plot(x1,y1,linestyle="dotted",color="g",linewidth='5')
plt.show()
```

```python
plt.plot(x1,y1,marker='^',ms='10',ls="dotted",color="g",linewidth='5')
plt.show()
```

```python
#multiple line in single graph
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
plt.plot(x,y,x1,y1)
plt.show()
```

```python
#set the title label (font)
font={'color':'green','size':'20'}
plt.title("line chart",fontdict=font,loc='left')			#loc=location
plt.xlabel('number1')
plt.ylabel('number2')
plt.plot(x1,y1)
plt.grid(axis='y',color="green",ls="--",linewidth='0.5') 		#grid in graph from y and x axis
plt.show()
```

```python
#1 graph in 2 graph (1 row 2 colums) subpot
plt.subplot(1,2,1)
plt.title("p1")
plt.plot(x,y)
plt.subplot(1,2,2) 			 # (2 row 2 column) 
plt.title("p2")
plt.plot(x1,y1)
plt.suptitle("super title")  		#supertitle of graphs
plt.show()
```

```python
#alpha for transpranecy size for the point
size=np.array([10,50])
plt.scatter(x1,y1,color='red')
plt.scatter(x,y,color='green' ,alpha=0.5)
plt.show()
```

```python
#ploting bar graph on x-axis
plt.bar(x1,y1)
plt.show()

#bar graph on y-axis
plt.barh(x1,y1)
plt.show()

#color and width  of bar 
plt.bar(x1,y1,color='r',width=10)
plt.show()
```

```python
x1=np.array([2,13,25,67])
y1=np.array([4,5,26,65])
plt.pie(x1)
plt.show()
```

```python
track_stud=np.array([20,30,40,20,10])
track_name=['Data','Mobile','Web','IOT','Cybersecurity']
ex=[0.2,0,0,0,0] #explode slice in the pie
col=['Red','pink','m','g','b'] #colors in pie chart
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=col) # startangle for starting angle of graph
plt.legend(title="Tracks")
plt.show()
```
